```python
import pandas as pd
import json
import seaborn as sns
import matplotlib.pyplot as plt
import pandas_profiling 

#with open('..\\tosca-metrics\\results\\industry_metric_results_5.json') as f:
with open('..\\tosca-metrics\\results\\example_metric_results_10.json') as f:
    data = json.load(f)

flat_dict = {}
for key, value in data.items():
    df = pd.io.json.json_normalize(value, sep='_')
    value = df.to_dict(orient='records')[0]
    flat_dict[key] = value

df = pd.DataFrame.from_dict(flat_dict, orient='index')
df = df = df[df.columns.drop(list(df.filter(regex='msg')))]
```

```python
df_nodetemplates = df_withdef[df_withdef['nn_count'] > 0]
len(df_nodetemplates)
```

```python
df_reltemplates = df_withdef[df_withdef['nr_count'] > 0]
len(df_reltemplates)
```
